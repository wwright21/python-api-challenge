```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create our url variable
url = "http://api.openweathermap.org/data/2.5/weather?"

# Creating empty lists in which we'll append stuff that's found
lat = []
lng = []
maxTemp = []
humidity = []
cloudiness = []
wndSpeed = []
country = []
date = []
foundCities = []

# Here is our for loop that will search for each city in our set and, if found, append the 
# stats to each list
for each in range(len(cities)):
    query_url = url + "appid=" + weather_api_key + "&q=" + cities[each] + '&units=imperial'
    response = requests.get(query_url).json()
    
    # Try to append our empty lists with data
    try:
        # this block of code will append the data we got from the response to the respective
        # empty lists above
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        maxTemp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wndSpeed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        foundCities.append(response['name'])
        
        # This print statement will display the name of the city and where we are in the 
        # count of cities
        print(f"{cities[each]} found, appending stats! | Record {each+1} out of {len(cities)}")
    except:
        # print that we were not successful
        print(f"{cities[each]} not found. On to the next...")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# To create the dataframe, we create a dictionary that will hold the values of our
# lists created in the above block of code
weather_dict = {
    "City": foundCities,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": maxTemp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wndSpeed,
    "Country": country,
    "Date": date
}

# Now we pass in the dictionary we just created in order to form the 'weather_data'
# dataframe
weather_data = pd.DataFrame(weather_dict)

# Set the index name to "City_ID"
weather_data.index.name = 'City_ID'

# Show the first 5 rows
weather_data.head()
```

```python
# Export the dataframe to a .csv
weather_data.to_csv('output_data/cities.csv')
```

## Linear Regression

```python
# In this section, we'll need to subsets of our weather_data dataframe: northern and southern 
# hemispheres. In this block of code, we'll create the 2 smaller dataframes needed for the 
# following section.

# Our northern dataset will only take rows whose latitudes are greater than 0
northern_df = weather_data.loc[weather_data['Lat'] > 0,:]

# Our southern dataset will take just the opposite
southern_df = weather_data.loc[weather_data['Lat'] < 0,:]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# First northern scatter plot. Similar to above scatter but with linear regression.
fig5 = northern_df.plot.scatter(x = 'Lat', y = 'Max Temp', grid=True, c='#87ceeb',
                         edgecolors='black',alpha=0.8);

# Now add the linear regression equation and line to plot
x_values = northern_df['Lat']
y_values = northern_df['Max Temp']

# using linregress from scipy.stats
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

# making a text version of our equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# draw our regression line
plt.plot(x_values,regress_values,"r-")

# annotate on our figure the line equation in red
plt.annotate(line_eq,(8,1),fontsize=12,color="red")
plt.annotate(f'r^2 value: {round(rvalue*rvalue, 3)}',(8, -10), fontsize=12, color="red")

# Now add the other fluff
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("City Latitude vs. Max Temperature - N. Hemisphere (04/19/22)")
plt.show()

# This final line will export the result as a .png
fig5.figure.savefig('output_data/Fig05_N Lat vs Temp.png')
```

#### Here we find a fairly strong correlation between the latitude of northern cities and the max temperature found, as evidenced by the visual fit of the red line and a fairly high r^2 value. As expected, max temperature falls as latitude increases (i.e., colder regions are found further away from the equator). Segregating northern and southern hemisphere cities is a more meaningful approach to analyzing these 2 variables
